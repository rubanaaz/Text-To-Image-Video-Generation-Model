```python
# === Import Libraries ===
import torch
from diffusers import StableDiffusionPipeline
from huggingface_hub import login
import gradio as gr
from getpass import getpass
from PIL import Image, ImageEnhance
import requests
import os
from io import BytesIO

# === Step 1: Hugging Face Authentication ===
print("🔐 Please enter your Hugging Face access token:")
token = getpass("Enter your Hugging Face access token: ")
login(token=token)

# === Step 2: Setup GPU ===
device = "cuda" if torch.cuda.is_available() else "cpu"
dtype = torch.float16 if device == "cuda" else torch.float32
print(f"✅ Using device: {device}")

# === Step 3: Load Stable Diffusion Model ===
pipe = StableDiffusionPipeline.from_pretrained(
    "runwayml/stable-diffusion-v1-5",
    torch_dtype=dtype
).to(device)

print("✅ Model loaded and ready!")

# === Step 4: Define Image Generation Function ===
def generate_image(prompt, style, num_images=1, size="512x512"):
    # Update generation settings based on style
    if style == "Photorealistic":
        pipe = StableDiffusionPipeline.from_pretrained("runwayml/stable-diffusion-v1-5", torch_dtype=dtype).to(device)
    elif style == "Painting":
        pipe = StableDiffusionPipeline.from_pretrained("runwayml/stable-diffusion-v1-5", torch_dtype=dtype).to(device)

    # Adjust the size based on user input
    width, height = map(int, size.split('x'))

    images = []
    for _ in range(num_images):
        image = pipe(prompt, height=height, width=width).images[0]
        images.append(image)

    return images

# === Step 5: Add Image Customization Features ===
def apply_filter(image, filter_type):
    if filter_type == "Sepia":
        sepia = ImageEnhance.Color(image).enhance(0.3)
        return sepia
    elif filter_type == "Black & White":
        bw = image.convert("L")
        return bw
    else:
        return image

# === Step 6: Upload Image to Imgur (Get URL) ===
def upload_to_imgur(image):
    # Upload the image to Imgur and get the link
    img_url = ""
    with BytesIO() as buffer:
        image.save(buffer, format="PNG")
        buffer.seek(0)
        img_data = buffer.read()

    headers = {"Authorization": "Client-ID your-client-id"}
    response = requests.post(
        "https://api.imgur.com/3/upload",
        headers=headers,
        files={"image": img_data}
    )
    if response.status_code == 200:
        img_url = response.json()["data"]["link"]
    return img_url

# === Step 7: Build Gradio UI ===
with gr.Blocks() as demo:
    gr.Markdown("## 🎨 Stable Diffusion Text-to-Image Generator with Extra Features")

    with gr.Row():
        prompt_input = gr.Textbox(label="Enter a prompt", placeholder="e.g., a fantasy castle in the clouds")
        style_input = gr.Dropdown(choices=["Photorealistic", "Painting", "Sketch"], label="Choose Style")
        num_images_input = gr.Slider(minimum=1, maximum=5, step=1, label="Number of Images", value=1)
        size_input = gr.Dropdown(choices=["512x512", "1024x1024"], label="Image Size", value="512x512")

    with gr.Row():
        generate_btn = gr.Button("Generate Image")

    with gr.Row():
        # Corrected: Use grid and height parameters directly in gr.Gallery
        image_output = gr.Gallery(label="Generated Images", show_label=True, grid=[2], height=200)

    with gr.Row():
        filter_input = gr.Dropdown(choices=["None", "Sepia", "Black & White"], label="Apply Filter")
        upload_btn = gr.Button("Upload to Imgur")
        img_url_output = gr.Textbox(label="Imgur Image URL")

    # === Step 8: Event Handling for Button Clicks ===
    def on_generate_click(prompt, style, num_images, size, filter_type):
        # Generate images
        images = generate_image(prompt, style, num_images, size)

        # Apply filter if selected
        images_with_filter = [apply_filter(img, filter_type) for img in images]

        return images_with_filter

    def on_upload_click(image):
        # Upload to Imgur and get URL
        img_url = upload_to_imgur(image)
        return img_url

    generate_btn.click(on_generate_click, inputs=[prompt_input, style_input, num_images_input, size_input, filter_input], outputs=[image_output])
    upload_btn.click(on_upload_click, inputs=[image_output], outputs=[img_url_output])

# === Step 9: Launch the App ===
demo.launch()
```

```python
import gradio as gr
from diffusers import StableDiffusionPipeline
import torch

# === 1) Load Model ===
model_id = "runwayml/stable-diffusion-v1-5"
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe = pipe.to("cuda")


# === 2) Image Generation Function ===
def ui_generate(prompt, style, num_images, resolution, filter_type, steps):
    # Apply the style selection
    if style == "Photorealistic":
        prompt = f"photorealistic {prompt}"
    elif style == "Painting":
        prompt = f"painting {prompt}"
    else:
        prompt = f"sketch {prompt}"

    # Apply the filter if selected
    if filter_type == "Sepia":
        filter_desc = "sepia toned"
    elif filter_type == "Black & White":
        filter_desc = "black and white"
    else:
        filter_desc = "colorful"

    prompt = f"{filter_desc} {prompt}"

    # Choose resolution
    if resolution == "512x512":
        width, height = 512, 512
    elif resolution == "512x640":
        width, height = 512, 640
    else:
        width, height = 640, 512

    # Generate images
    images = pipe(prompt, num_inference_steps=steps, width=width, height=height, num_images_per_prompt=num_images).images

    return images

# === 3) Gradio UI Components ===
with gr.Blocks() as demo:
    gr.Markdown("## 🎨 SD + Upscaler Text-to-Image")

    with gr.Row():
        prompt_in = gr.Textbox(label="Prompt", placeholder="e.g. portrait of a girl")
        style_in  = gr.Dropdown(["Photorealistic", "Painting", "Sketch"], label="Style")
        num_in    = gr.Slider(1, 5, 1, label="# Images")
        size_in   = gr.Dropdown(["512x512", "512x640", "640x512"], label="Res (W×H)")
        steps_in  = gr.Slider(20, 100, 80, label="Steps")
        filter_in = gr.Dropdown(["None", "Sepia", "Black & White"], label="Filter")

    btn = gr.Button("Generate")

    # Create the gallery without style() method
    gallery = gr.Gallery(
        label="Output",
        columns=2,
        show_label=True,
        height="auto"
    )

    btn.click(
        ui_generate,
        inputs=[prompt_in, style_in, num_in, size_in, filter_in, steps_in],
        outputs=[gallery]
    )

    demo.launch()
```

```python
import gradio as gr
from diffusers import StableDiffusionPipeline
import torch

# Load Stable Diffusion pipeline
model_id = "runwayml/stable-diffusion-v1-5"
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe.to("cuda")

# Image generation function
def generate_image(prompt, style, resolution):
    if style == "Realistic":
        prompt = f"photorealistic, {prompt}"
    elif style == "Cartoon":
        prompt = f"cartoon style, {prompt}"
    elif style == "Fantasy":
        prompt = f"fantasy art, {prompt}"

    width, height = map(int, resolution.lower().split('x'))
    image = pipe(prompt, height=height, width=width).images[0]
    return image

# Gradio Interface
with gr.Blocks(css="""
#sidebar {
    background-color: #1f1f2e;
    height: 100vh;
    padding: 20px;
    color: white;
}
.sidebar-button {
    background-color: transparent;
    border: none;
    font-size: 16px;
    text-align: left;
    color: white;
    margin-bottom: 10px;
}
#title {
    color: white;
    font-size: 24px;
    font-weight: bold;
    text-align: center;
    margin-bottom: 20px;
}
""") as demo:

    with gr.Row():
        with gr.Column(scale=1, min_width=200, elem_id="sidebar"):
            gr.Markdown("ImageVisionAI", elem_id="title")
            gr.Button("🖌️ Create", elem_classes="sidebar-button")
            gr.Button("🖼️ Gallery", elem_classes="sidebar-button")
            gr.Button("⚙️ Settings", elem_classes="sidebar-button")
            gr.Button("❓ Help", elem_classes="sidebar-button")

        with gr.Column(scale=4):
            gr.Markdown("## AI Content Generator")
            with gr.Tabs():
                with gr.TabItem("Image Generation"):
                    prompt = gr.Textbox(label="Prompt", placeholder="a boy sitting on the top of mountain")
                    style = gr.Dropdown(["Realistic", "Cartoon", "Fantasy"], label="Style", value="Realistic")
                    resolution = gr.Dropdown(["512x512", "768x768", "1024x1024"], label="Resolution", value="512x512")
                    generate_btn = gr.Button("Generate 🔥")
                    output_image = gr.Image(label="Generated Image")
                    generate_btn.click(fn=generate_image, inputs=[prompt, style, resolution], outputs=output_image)

                with gr.TabItem("Video Generation"):
                    gr.Markdown("🚧 Coming soon...")

    gr.Markdown("### Generated Results")

# Launch the app
demo.launch()
```
